```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def central_limit_theorem(sample_size, num_samples):
  sample_means = []
  for _ in range(num_samples):
    sample = np.random.uniform(0, 1, sample_size)
    sample_means.append(np.mean(sample))
  return sample_means
```

```python
sample_means_10 = central_limit_theorem(10, 1000)
sample_means_50 = central_limit_theorem(50, 1000)
sample_means_100 = central_limit_theorem(100, 1000)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(sample_means_10, bins=30, density=True, alpha=0.5, label='Sample Size = 10')
plt.hist(sample_means_50, bins=30, density=True, alpha=0.5, label='Sample Size = 50')
plt.hist(sample_means_100, bins=30, density=True, alpha=0.5, label='Sample Size = 100')

# Add a normal distribution curve for comparison
x = np.linspace(0, 1, 100)
plt.plot(x, norm.pdf(x, 0.5, 1/np.sqrt(12*100)), 'r-', label='Normal Distribution')

plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title('Central Limit Theorem')
plt.legend()
plt.show()
```
